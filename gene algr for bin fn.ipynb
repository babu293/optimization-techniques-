```python
import random
```

```python
def binary_function(x):
    return sum(x)  # Example function: sum of binary digits
```

```python
# Genetic Algorithm parameters
population_size = 50
chromosome_length = 10
mutation_rate = 0.01
num_generations = 100
```

```python
# Generate initial population
population = []
for _ in range(population_size):
    chromosome = [random.randint(0, 1) for _ in range(chromosome_length)]
    population.append(chromosome)
```

```python
# Main Genetic Algorithm loop
for generation in range(num_generations):
    print(f"Generation {generation + 1}")

    # Evaluate fitness of each individual in the population
    fitness_scores = [binary_function(chromosome) for chromosome in population]

    # Find the best individual
    best_individual = population[fitness_scores.index(max(fitness_scores))]
    print("Best Individual:", best_individual, "Fitness:", max(fitness_scores))

    # Create a new population through selection, crossover, and mutation
    new_population = []
    while len(new_population) < population_size:
        # Selection
        parent1 = random.choices(population, weights=fitness_scores)[0]
        parent2 = random.choices(population, weights=fitness_scores)[0]

        # Crossover
        crossover_point = random.randint(1, chromosome_length - 1)
        child = parent1[:crossover_point] + parent2[crossover_point:]

        # Mutation
        for i in range(chromosome_length):
            if random.random() < mutation_rate:
                child[i] = 1 - child[i]  # Flip the bit

        new_population.append(child)

    # Replace the old population with the new population
    population = new_population
```

```python
# Final best individual
best_individual = population[fitness_scores.index(max(fitness_scores))]
print("Final Best Individual:", best_individual, "Fitness:", max(fitness_scores))
```
